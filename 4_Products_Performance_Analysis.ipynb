## üìÇ Load the Dataset  
* We begin by loading the sales dataset to explore its structure and contents.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("drive/MyDrive/Colab Notebooks/Data.csv")

# Display first few rows
print(df.head())
```

#üìä Analyzing Product Contribution to Sales  

###To evaluate **product performance**, we analyze total sales across different product lines.  
This helps in understanding:  

####‚úÖ Which **product lines drive the most revenue** üìà  
####‚úÖ Which **products contribute the least to overall sales** üìâ  

####By visualizing this data, businesses can focus on **high-performing products** and improve strategies for the **low-performing ones**.

```python
# Analyzing Products Contribution to Sales

# Group by Product Line and sum total sales
Top_Products = df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False) / 1e5

# Set figure size
plt.figure(figsize=(10, 5))

# Create a bar plot
Top_Products.plot(kind="bar", color="teal")

# Add title and labels
plt.title("Top Revenue-Generating Product Categories", fontsize=14, fontweight="bold")
plt.xlabel("Product Line")
plt.ylabel("Total Sales (in Lakhs)")

# Display the plot
plt.show()
```

# üìä Sales Distribution Across Different Product Lines  

#### Understanding the **sales distribution** across different product lines helps in identifying which categories generate the highest revenue and which contribute the least.  

## By visualizing this distribution, we can:  

####‚úÖ Determine **which product lines dominate sales** üìà  
####‚úÖ Identify **less contributing product lines** üìâ  
####‚úÖ Help businesses focus on **high-performing categories** while improving weaker ones  

####We use **pie charts and bar charts** to represent the proportion of total sales contributed by each product line. This provides a **clear visual breakdown** of sales across categories.

```python
# Group Sales by Product Line
Sales_Distribution=df.groupby("PRODUCTLINE")["SALES"].sum()

# Plot Pie Chart for Distribution
Sales_Distribution.plot(kind="pie",autopct="%1.1f%%", colors=["#ff9999","#66b3ff","#99ff99","#ffcc99","#c2c2f0","#ffb3e6","#c4e17f"])

# Add Title
plt.title("Sales DIrtribution Across Product Lines")

# Display the plot
plt.ylabel("")  # Remove default y-label
plt.show()
```
