## ðŸ“‚ Load the Dataset  
* We begin by loading the sales dataset to explore its structure and contents.

```python
import pandas as pd               # Data manipulation
import matplotlib.pyplot as plt   # Visualization
import seaborn as sns             # Advanced plots

df=pd.read_csv("drive/MyDrive/Colab Notebooks/Data.csv")
print(df.head())
```

#ðŸ”¥Order Status Breakdown
####âœ…You can calculate the percentage of orders in each STATUS (e.g., Shipped, Cancelled, On Hold) and plot a pie chart for a quick overview.

```python
# Order Status Breakdown (Bar Graph)
status_counts = df['STATUS'].value_counts()
status_percentage = (status_counts / status_counts.sum()) * 100  # Convert to percentage

# Convert to DataFrame for plotting
status_df = pd.DataFrame({'STATUS': status_percentage.index, 'PERCENTAGE': status_percentage.values})

# Plot: Order Status Breakdown (Fixed Bar Chart)
plt.figure(figsize=(10,6))
sns.barplot(x='STATUS', y='PERCENTAGE', data=status_df, hue='STATUS', dodge=False, palette='Set2', legend=False)

plt.title('Order Status Breakdown')
plt.xlabel('Order Status')
plt.ylabel('Percentage of Orders (%)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

#ðŸ”¥Sales Contribution by Deal Size
####âœ…You can calculate the total SALES for each DEALSIZE category (Small, Medium, Large) and plot a bar chart to show the contribution of each deal size to the total sales.

```python
# âœ… Sales Contribution by Deal Size (Bar Chart)
deal_size_sales = df.groupby('DEALSIZE')['SALES'].sum().reset_index()
deal_size_sales['SALES'] = deal_size_sales['SALES'] / 1e5

# Plot: Sales Contribution by Deal Size
plt.figure(figsize=(10,6))
sns.barplot(x='DEALSIZE', y='SALES',hue='DEALSIZE',data=deal_size_sales, palette='coolwarm')
plt.title('Sales Contribution by Deal Size')
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel('Deal Size')
plt.ylabel('Total Sales (in Lakhs)')
plt.show()
```
